    Korzystając z przygotowanych do laboratorium plików wczytuję dane z pliku pobieranego przez program i wyświetlam je na ekranie. Plik zawiera dane o huraganach w USA w latach 2005 - 2015
    
    

```python
import pandas as pd
import matplotlib.pyplot as plt

import urllib.request
import os

plik = 'hurricanes.csv'
URL = 'https://people.sc.fsu.edu/~jburkardt/data/csv/' + plik
if not os.path.isfile(plik):
    print('Pobieram plik ', URL)
    urllib.request.urlretrieve(URL, plik)
    print('Pobrano.')
else:
    print('Plik już jest na dysku.')

df = pd.read_csv(plik)
print(df)
```

Tera przedstawmy te dane na wykresie z pomocą następującej funkcji. Właściwie przedstawiam te dane dla każdego roku osobno.

```python
plots = []

def createPlot(year):
    plot = df.plot(title='Huragany w USA w ' + year, x='Month', y=year, kind='scatter')
    plot.set_ylabel('Ilość huraganów')
    plot.set_xlabel("Miesiąc")
    
    plots.append(plot)
    
createPlot(' "2005"')
createPlot(' "2006"')
createPlot(' "2007"')
createPlot(' "2008"')
createPlot(' "2009"')
createPlot(' "2010"')
createPlot(' "2011"')
createPlot(' "2012"')
createPlot(' "2013"')
createPlot(' "2014"')
createPlot(' "2015"')

plt.show()
        
```

Czas dodać średnią na każdym wykresie. Posłuży nam do tego następujący kod:

```python
avr = df[[' "2005"']].sum() / df[[' "2005"']].count()
avr.plot()
```
